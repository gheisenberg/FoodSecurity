```python
input_dir = "/mnt/datadisk/data/Projects/water/inputs/"
moz_preds_dir = "/mnt/datadisk/data/Projects/water/inputs/Moz_preds/"
gaul_f = f"{input_dir}gaul_Moz_2015-2014_adm2.shp"
```

```python
import geopandas as gpd
import os
import pandas as pd

# Read the shapefile
gaul_df = gpd.read_file(gaul_f)
gaul_df
```

```python

# Initialize an empty dataframe to store the merged data
merged_df = gpd.GeoDataFrame()

# Loop through all the prediction files
for file in os.listdir(moz_preds_dir):
    if file.endswith('.csv'):  # check if the file is a CSV file
        # Read the prediction file
        pred_df = pd.read_csv(os.path.join(moz_preds_dir, file))
        
        cols = [col for col in pred_df.columns if col not in gaul_df.columns]
        if 'adm2_gaul' not in cols:
            cols.append('adm2_gaul')
        
        # If 'adm2_gaul' already exists in gaul_df, drop it
        if 'adm2_gaul' in gaul_df.columns:
            gaul_df = gaul_df.drop(columns=['adm2_gaul'])

                
        # Merge the prediction file with the shapefile
        merged = gaul_df.merge(pred_df[cols], how='left', left_on='adm2_name', right_on='adm2_gaul')  # replace 'id' with the column name that is common between the two dataframes
        # write to shapefile
        file_new = file.replace('.csv', '.shp')
        file_new = file_new.replace(':', '')
        merged.to_file(f"{moz_preds_dir}{file_new}")
gaul_df
```
